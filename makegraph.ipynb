For each row in unmatchbio, find the correspongding row in pkg

gene: direct match

mesh&disease: use MESH-MONDO.cv, mondo/grouped(some mesh id don't have a matched mondo id), https://github.com/natacourby/Disease_ontologies_for_knowledge_graphs

omim&disease:

mesh&chemical: drugbank



```python
import re
updated_rows = []
for index, row in unmatchbio.iterrows():
    if row['source'] == 'ncbi_gene':
        match = pkg[(pkg['x_id'] == row['id']) & (pkg['x_source'] == 'NCBI')]
        if not match.empty:
            # Add the target DataFrame's columns to the row
            for col in match.columns:
                row[col] = match[col].values[0]
        else: 
            row['x_index'] = 'NaN'
            row['x_name'] = row['name']
            row['x_id'] = row['id']
            row['x_type'] = 'gene/protein'
            row['x_source'] = 'NCBI'

    elif row['source'] == 'ncbi_mesh' and row['type'] == 'Disease': # map mesh disease to MONDO
        # Check if the id is in meshMONDO['curie_id']
        match = meshMONDO[meshMONDO['xref'].str.lower() == row['id'].lower()]
        if not match.empty:
            # Get the corresponding meshMONDO['mapped_curie']
            mapped_curie = match['cls'].values[0]
            # Extract the digits from the right until zero
            match_digits = re.search(r'0*(\d+)', mapped_curie.split(':')[-1])
            extracted_digits = match_digits.group(1)
            # MONDO entities that are not in PrimeKG
            row['x_index'] = 'NaN'
            row['x_name'] = row['name']
            row['x_id'] = extracted_digits
            row['x_type'] = 'disease'
            row['x_source'] = 'MONDO'
            # x_source is MONDO
            pkg_match = pkg[(pkg['x_id'].astype(str) == extracted_digits) & (pkg['x_source'] == 'MONDO')]
            if not pkg_match.empty:
                # Add the target DataFrame's columns to the row
                for col in pkg_match.columns:
                    row[col] = pkg_match[col].values[0]
            # x_source is MONDO_grouped
            pkg['x_id_split'] = pkg['x_id'].apply(lambda x: x.split('_'))
            pkg_match1 = pkg[(pkg['x_id_split'].apply(lambda x: extracted_digits in x)) & (pkg['x_source'] == 'MONDO_grouped')]
            if not pkg_match1.empty:
                # Add the target DataFrame's columns to the row
                for col in pkg_match1.columns:
                    row[col] = pkg_match1[col].values[0]
            # Drop the temporary 'x_id_split' column
            pkg.drop(columns=['x_id_split'], inplace=True)
        else:
            row['x_index'] = 'NaN'
            row['x_name'] = row['name']
            row['x_id'] = row['id']
            row['x_type'] = 'disease'
            row['x_source'] = 'ncbi_mesh'
    elif row['source'] == 'omim' and row['type'] == 'Disease': # map omim disease to MONDO
        pass
    elif row['source'] == 'ncbi_mesh' and row['type'] == 'Chemical': # map mesh chemical to drugbank
        id = row['id'].split(':')[1]
        match = meshDB[meshDB['Mesh_ID'] == id]
        if not match.empty:
            mapped_curie = match['drugBank_ID'].values[0]
            # DB entities that are not in PrimeKG
            row['x_index'] = 'NaN'
            row['x_name'] = row['name']
            row['x_id'] = mapped_curie
            row['x_type'] = 'drug'
            row['x_source'] = 'DrugBank'
            pkg_match = pkg[(pkg['x_id'].astype(str) == mapped_curie) & (pkg['x_source'] == 'DrugBank')]
            if not pkg_match.empty:
                # Add the target DataFrame's columns to the row
                for col in pkg_match.columns:
                    row[col] = pkg_match[col].values[0]
        else:
            row['x_index'] = 'NaN'
            row['x_name'] = row['name']
            row['x_id'] = row['id']
            row['x_type'] = 'drug'
            row['x_source'] = 'ncbi_mesh'

    # match for other types that are not in PrimeKG
    else: 
        row['x_index'] = 'NaN'
        row['x_name'] = row['name']
        row['x_id'] = row['id']
        row['x_type'] = row['type']
        row['x_source'] = row['source']

    updated_rows.append(row)

updated_df = pd.DataFrame(updated_rows)
# drop columns 'x_id'
updated_df.drop(columns=['x_id_split'], inplace=True)
updated_df
```

```python
def get_hops(df, start_node, max_hops=6):
    visited = set()  # Set to track visited nodes
    current_nodes = set([start_node])  # Start with the initial node

    for _ in range(max_hops):
        new_nodes = set()
        # Find all rows where the current set of nodes are mentioned in x_name or y_name
        query = df[(df['x_name'].isin(current_nodes)) | (df['y_name'].isin(current_nodes))]
        # Add all connected nodes to the new set
        new_nodes.update(query['x_name'].tolist())
        new_nodes.update(query['y_name'].tolist())
        # Update the visited nodes
        visited.update(current_nodes)
        # Update current_nodes with new nodes that haven't been visited
        current_nodes = new_nodes - visited

    return df[(df['x_name'].isin(visited)) | (df['y_name'].isin(visited))]
```

```python
# Get 6 hops from 'Alzheimer disease'
alzheimer_hops = get_hops(primekg, 'Alzheimer disease', max_hops=4)

# Save the sub-graph to a new CSV file
alzheimer_hops.to_csv('data/alzheimer_sub_graph4.csv', index=False)
```
